## Activity
Search for a weather API. Get data from the last seven days. You must include:

- Temperature
- Rain (mm)

Save the data in a **CSV file**.

```python
import requests as rq
import pandas as pd
from datetime import datetime, timedelta


# Method to make daily frame
def makeTable(timeData, targetCity):
    # Daily means last 7 days
    daily = timeData['daily']
    df = pd.DataFrame({
        "city": targetCity,
        "day": daily['time'],
        "temp_max": daily['temperature_2m_max'],
        "rain": daily['precipitation_sum'],
        "wind": daily['windspeed_10m_max']
    })
    df.to_csv("tableDaily.csv", index=False)

# Method to print table
def printTable():
  df = pd.read_csv("tableDaily.csv")
  display(df)

# Varible to guard lat and lon of insert citys
# Start with: Campinas, Indaiatuba and Hortolândia
coordinates = [
    (-22.9056, -47.0608),
    (-23.0900, -47.2178),
    (-22.85833, -47.22000),
]

# Today
end_date = datetime.today()
start_date = end_date - timedelta(days=7)
# format to string (API format)
end_date_str = end_date.strftime("%Y-%m-%d")
start_date_str = start_date.strftime("%Y-%m-%d")

while True:
  print("====== Daily city's weather ======")
  lat = float(input('Type your latitude: '))
  lon = float(input('Type your longitude: '))
  coordinates.append((lat, lon))

  op = input("Do you wanna get information of another city? (Y/N)\n> ").upper()
  if op != 'Y':
    break
  else:
    continue

# Target url for request
url_weather = 'https://api.open-meteo.com/v1/forecast'

# Get nominatim API
url_location = "https://nominatim.openstreetmap.org/reverse"

# Header required by Nominatim API for requests
headers = {
    "User-Agent": "meu_app_teste"
}

# Results for request
dfs = []

for lat, lon in coordinates:
  # Parameters sent to the API request Weather
  params = {
      # Geographic coordinates for API request
      "latitude": lat,
      "longitude": lon,
      # Daily weather variables returned by API
      "daily": "temperature_2m_max,precipitation_sum,windspeed_10m_max",
      "start_date": start_date_str,
      "end_date": end_date_str,
      "timezone": "auto"
  }
  # Parameters sent to the API request Location
  param_geo = {
    "lat": lat,
    "lon": lon,
    "format": "json"
  }
  # Send request to the API weather with the parameters
  response_weather = rq.get(url_weather, params=params)
  # Send request to the API location with the parameters
  response_location = rq.get(url_location, params=param_geo, headers=headers)
  if response_weather.status_code == 200 and response_location.status_code == 200:
    # Data for weather and location
    data_weather = response_weather.json()
    data_location = response_location.json()

    #Variable to get target city in API retunr in a object
    endereco = data_location.get("address", {})

    # Get the target city in object
    city = endereco.get("city") or endereco.get("town") or endereco.get("village")

    # Make table with weather and the target city
    df = makeTable(data_weather, city)
    dfs.append(df)

    # Show Table with all citys on:
    printTable()

  else:
    print("API ERROR...")

```
